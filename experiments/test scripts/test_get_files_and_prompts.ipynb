## Utility function : extract skill code

```python
def extract_skill_name(code_file,file_name:str):
    """
    Extracts the skill code from the file name
    """

    # Check if robust in string or not
    robust = "robust" in file_name
    name = file_name[:-11] if robust else file_name[:-4]
    # Retrive the skill code
    skill_code = name.split("_")[-3]
    print(f"Skill code : {skill_code}")
    # Retrieve the skill code
    code = code_file.loc[code_file["code"]==int(skill_code), "skill"]
    # Check if the code is not empty
    if not code.empty: 
        code = code.iloc[0]
    else: 
        code = -1

    return code
```
